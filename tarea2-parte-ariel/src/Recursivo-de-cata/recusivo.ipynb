```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('recAriel.csv')
x= df['n']
y=df['t']
```

```python
plt.ylabel("Tiempo (s)")
plt.xlabel("Numero de Fibonacci")
plt.title("Fibonacci Recursivo")
plt.plot(x, y,marker=".")
plt.savefig("RecAriel.png")
```
